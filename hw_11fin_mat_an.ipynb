## 1. Решить уравнения:

## 1. y' − y = 2x − 3;

```python
import sympy as sp

# Определение переменных
x, C = sp.symbols('x C')
y = sp.Function('y')(x)

# Определение дифференциального уравнения
diff_eq = sp.Eq(y.diff(x) - y, 2*x - 3)

# Решение уравнения
solution = sp.dsolve(diff_eq, y)
print(f"Общее решение уравнения: {solution}")
```

## 2. x^2 y' + xy + 1 =0

```python
# Определение переменных и функции
x = sp.Symbol('x')
y = sp.Function('y')(x)

# Определение дифференциального уравнения
diff_eq = sp.Eq(x**2 * y.diff(x) + x * y + 1, 0)

# Решение уравнения
solution = sp.dsolve(diff_eq, y)
print(f"Общее решение уравнения: {solution}")
```

## 2. Исследовать ряд на сходимость:

## 1). ∑ (n + 2)/(n^2 + n + 1) n = [1; ∞]



```python
# Определение переменных
n = sp.Symbol('n', integer=True)

# Определение общего члена ряда
a_n = (n + 2) / (n**2 + n + 1)

# Проверка поведения общего члена ряда при больших n
limit_behavior = sp.limit(a_n, n, sp.oo)
print(f"Предел общего члена ряда при n → ∞: {limit_behavior}")

# Анализ ряда
# Поскольку ряд может быть сравним с гармоническим рядом, проверим это
harmonic_comparison = a_n / (1 / n)
limit_comparison = sp.limit(harmonic_comparison, n, sp.oo)
print(f"Предел отношения общего члена к 1/n: {limit_comparison}")

# Можно проверить несколько первых членов ряда для численной проверки
N = 1000  # Количество членов ряда для суммирования
numerical_sum = sum([a_n.subs(n, i) for i in range(1, N+1)])
print(f"Численная сумма первых {N} членов ряда: {numerical_sum}")
```

## 2) ∑ (n^n)/n! n = [1; ∞]

```python
# Определение переменной
n = sp.Symbol('n', integer=True)

# Определение общего члена ряда
a_n = n / sp.factorial(n)

# Вычисление суммы ряда
# Используем встроенную функцию sum() в sympy для вычисления суммы ряда
sum_series = sp.summation(a_n, (n, 1, sp.oo))
print(f"Сумма ряда: {sum_series}")

# Проверка сходимости ряда
# Мы также можем использовать функцию limit для анализа поведения общего члена
limit_behavior = sp.limit(a_n, n, sp.oo)
print(f"Предел общего члена ряда при n → ∞: {limit_behavior}")
```

## 3) ∑ n * (1 -1/n)^(n^2); n = [1; ∞]

```python
import sympy as sp

# Определение переменной
n = sp.Symbol('n', real=True, positive=True)

# Определение общего члена ряда
a_n = n * (1 - 1 / n)**(n**2)

# Проверка предела общего члена ряда при n → ∞
limit_behavior = sp.limit(a_n, n, sp.oo)
print(f"Предел общего члена ряда при n → ∞: {limit_behavior}")

# Анализ сходимости ряда
# Для этого можно использовать тесты сравнения
# Используем тест на абсолютную сходимость

# Определим общий член ряда для абсолютной сходимости
a_n_absolute = sp.Abs(a_n)
absolute_convergence = sp.summation(a_n_absolute, (n, 1, sp.oo))

# Численная проверка
import numpy as np

def numerical_sum(N):
    return np.sum([i * (1 - 1 / i)**(i**2) for i in range(1, N + 1)])

N = 100  # Количество членов для проверки
numerical_sum_value = numerical_sum(N)
print(f"Численная сумма первых {N} членов ряда: {numerical_sum_value}")
```

## 5). ∑ (x-2)^(n+1)/ 3^n (n + 2); n = [1; ∞]

```python
import sympy as sp
import numpy as np

# Определение переменных
n, x = sp.symbols('n x', real=True, positive=True)

# Определение общего члена ряда
def a_n_func(x_val, n_val):
    return (x_val - 2)**(n_val + 1) / (3**n_val * (n_val + 2))

# Функция для вычисления отношения a_{n+1} / a_n
def ratio_test(x_val, n_val):
    a_n = a_n_func(x_val, n_val)
    a_n_plus_1 = a_n_func(x_val, n_val + 1)
    return np.abs(a_n_plus_1 / a_n)

# Численное вычисление предела отношения a_{n+1} / a_n
x_val = 0  # Значение x для проверки
N = 1000  # Количество членов для проверки
ratios = [ratio_test(x_val, i) for i in range(1, N)]

# Найдем предел отношения
limit_ratio = np.mean(ratios[-50:])  # Среднее значение последних нескольких отношений
radius_of_convergence = 1 / limit_ratio
print(f"Численное приближение радиуса сходимости: {radius_of_convergence}")

# Проверка предела общего члена при n → ∞
def numerical_limit(x_val):
    return np.mean([a_n_func(x_val, i) for i in range(1, N)])

limit_behavior = numerical_limit(x_val)
print(f"Численное приближение предела общего члена при n → ∞: {limit_behavior}")

# Численная проверка для конкретного значения x
def numerical_sum(x_val, N):
    return np.sum([a_n_func(x_val, i) for i in range(1, N + 1)])

# Пример для x = 0 и N = 100
numerical_sum_value = numerical_sum(x_val, N)
print(f"Численная сумма первых {N} членов ряда для x = {x_val}: {numerical_sum_value}")
```
