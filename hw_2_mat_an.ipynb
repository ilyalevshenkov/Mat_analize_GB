## Задание №1

Пусть {xn} = n ^ (−1) ^ n.

Доказать, что последовательность {xn}:

а) неограниченная;

б) не является бесконечно большой.

Для того чтобы доказать утверждения, разобьем задачу на два пункта.

### а) Неограниченность последовательности {xn}

#### Четные и нечетные ( n ):

- Если \( n \) чётное, то (-1)^n = 1, и тогда ( x_n = n^1 = n.
- Если \( n \) нечётное, то \( (-1)^n = -1 \), и тогда \( x_n = n^{-1} = \frac{1}{n} \).

Теперь проверим неограниченность последовательности.

- Для четных \( n \), элементы последовательности \( x_n = n \) возрастают без ограничения. То есть, для любой сколь угодно большой величины \( M \), существует четное \( n \) такое, что \( x_n = n > M \).
  
Таким образом, последовательность \( \{x_n\} \) неограниченна, так как хотя бы часть элементов (те, что соответствуют четным значениям \( n \)) растет без ограничения.

### б) Не является бесконечно большой

Для того чтобы последовательность была бесконечно большой, необходимо, чтобы для всех значений \( n \) элементы последовательности \( x_n \) стремились к бесконечности при \( n \to \infty \).

- Если \( n \) нечётное, то \( x_n = \frac{1}{n} \), и этот элемент стремится к нулю при \( n \to \infty \).
  
Следовательно, поскольку часть элементов последовательности (те, что соответствуют нечётным значениям \( n \)) стремятся к нулю, последовательность \( \{x_n\} \) не является бесконечно большой.

### Ответ:
а) Последовательность \( \{x_n\} \) неограниченна.

б) Последовательность \( \{x_n\} \) не является бесконечно большой.

```python
import numpy as np
import matplotlib.pyplot as plt

# Зададим количество элементов последовательности
N = 1000

# Определим последовательность xn
n_values = np.arange(1, N+1)
x_values = np.array([n if n % 2 == 0 else 1/n for n in n_values])

# Разделим на четные и нечетные
even_indices = n_values % 2 == 0
odd_indices = n_values % 2 != 0

# Построим график последовательности
plt.figure(figsize=(10, 6))
plt.plot(n_values[even_indices], x_values[even_indices], 'bo-', label="Четные n (x_n = n)")
plt.plot(n_values[odd_indices], x_values[odd_indices], 'ro-', label="Нечетные n (x_n = 1/n)")
plt.title("График последовательности x_n = n^{(-1)^n}")
plt.xlabel("n")
plt.ylabel("x_n")
plt.legend()
plt.grid(True)
plt.show()

# Выведем несколько значений для анализа
print(f"Первые 10 элементов последовательности: {x_values[:10]}")
```

## Задача №2.
Доказать, что последовательность {sin n} расходится.

Чтобы доказать, что последовательность {sin n} расходится, используем понятие расходимости последовательности в математическом анализе. Рассмотрим несколько важных шагов и аргументов.


1. **Периодичность и плотность**:
   Функция sin(x) является периодической с периодом 2pi, то есть для любого числа x, выполняется:
  
   sin(x + 2\pi) = sin(x).
  
   Однако, для последовательности {sin n }, где \( n \) принимает только целые значения, не существует очевидного периода, потому что последовательность не проходит через целые кратные периодов \( 2\pi \).

2. **Плотность синуса на интервале**:
   Известно, что множество чисел ( { n \mod 2\pi \} ), где ( n ) — целое число, является плотным на отрезке [0, 2\pi] . Это означает, что значения последовательности ( sin n ) не сосредотачиваются вокруг одного числа, а могут приближаться к любому значению в диапазоне от (-1) до (1).

3. **Аргумент против сходимости**:
   Пусть последовательность { sin n } сходится к некоторому числу ( L ). Тогда, по определению сходимости последовательности, для любого ( epsilon > 0 ) существует натуральное число  N , такое что для всех ( n > N ) выполняется:

    sin n - L < epsilon

   Однако, поскольку значения (sin n ) распределены хаотично и могут приближаться к любому значению на интервале  [-1, 1], невозможно найти такое число ( L ), к которому последовательность будет стремиться, и это приводит к противоречию.

### Заключение:

Последовательность {sin n } не сходится ни к одному конкретному числу, поскольку её значения не стабилизируются, а продолжают колебаться между (-1) и \(1) по мере роста ( n ). Следовательно, последовательность { sin n } **расходится**.

```python
# Задаем количество элементов последовательности
N = 1000

# Генерируем последовательность sin(n) для n от 1 до N
n_values = np.arange(1, N + 1)
sin_values = np.sin(n_values)

# Построим график последовательности {sin(n)}
plt.figure(figsize=(10, 6))
plt.scatter(n_values, sin_values, color='blue', s=10, label='sin(n)')
plt.axhline(y=0, color='r', linestyle='--')
plt.title("График последовательности {sin(n)} для n от 1 до 1000")
plt.xlabel("n")
plt.ylabel("sin(n)")
plt.legend()
plt.grid(True)
plt.show()

# Выведем несколько первых элементов последовательности для анализа
print(f"Первые 10 значений sin(n): {sin_values[:10]}")

# Дополнительно построим гистограмму, чтобы показать плотность значений sin(n)
plt.figure(figsize=(10, 6))
plt.hist(sin_values, bins=30, color='blue', alpha=0.7, edgecolor='black')
plt.title("Гистограмма распределения значений sin(n)")
plt.xlabel("Значения sin(n)")
plt.ylabel("Частота")
plt.grid(True)
plt.show()
```

```python
# Задаем количество элементов последовательности
N = 1000

# Генерируем последовательность sin(n) для n от 1 до N
n_values = np.arange(1, N + 1)
sin_values = np.sin(n_values)

# Построим график последовательности {sin(n)} для выборочных n
plt.figure(figsize=(12, 6))
plt.plot(n_values, sin_values, color='green', marker='o', linestyle='-', markersize=2, label='sin(n)')
plt.axhline(y=0, color='r', linestyle='--', label='y = 0')
plt.title("График последовательности {sin(n)} для n от 1 до 1000")
plt.xlabel("n")
plt.ylabel("sin(n)")
plt.legend()
plt.grid(True)
plt.show()

# Дополнительно анализируем плотность через выборочное распределение по модулю 2*pi
plt.figure(figsize=(12, 6))
mod_2pi_values = n_values % (2 * np.pi)
sin_mod_values = np.sin(mod_2pi_values)

plt.scatter(mod_2pi_values, sin_values, color='purple', alpha=0.5, s=10, label='sin(n mod 2π)')
plt.title("Распределение значений sin(n mod 2π)")
plt.xlabel("n % 2π")
plt.ylabel("sin(n mod 2π)")
plt.grid(True)
plt.legend()
plt.show()

# Рассмотрим несколько числовых значений для анализа
print(f"Первые 10 значений sin(n): {sin_values[:10]}")
print(f"Некоторые значения sin(n mod 2π): {sin_mod_values[:10]}")

# Рассчитаем статистику плотности значений на интервале [-1, 1]
min_val = np.min(sin_values)
max_val = np.max(sin_values)
mean_val = np.mean(sin_values)

print(f"Минимальное значение sin(n): {min_val}")
print(f"Максимальное значение sin(n): {max_val}")
print(f"Среднее значение sin(n): {mean_val}")
```

### Результат:
График показывает, что последовательность стремится к бесконечности по мере увеличения \( n \).

можно написать код на Python, который покажет, как это выражение ведет себя при больших значениях
n. В этом коде мы будем вычислять значения последовательности для разных
n и наблюдать их приближение к пределу.

```python
#import numpy as np
#import matplotlib.pyplot as plt

# Задаем диапазон значений n
n_values = np.arange(1, 1000000, 10000)  # Увеличиваем диапазон для точности
sequence_values = np.sqrt(n_values**2 + n_values) - n_values

# Выводим результаты для больших значений n
print(f"Значения последовательности для больших n:")
for n, value in zip(n_values[-10:], sequence_values[-10:]):
    print(f"n = {n}, sqrt(n^2 + n) - n = {value}")

# Вычислим и выведем значение для самого большого n
largest_n = n_values[-1]
largest_value = np.sqrt(largest_n**2 + largest_n) - largest_n
print(f"\nЗначение последовательности для самого большого n ({largest_n}): {largest_value}")

# Cтроим график
plt.figure(figsize=(10, 6))
plt.plot(n_values, sequence_values, color='blue', label='sqrt(n^2 + n) - n')
plt.axhline(y=1/2, color='r', linestyle='--', label='Ожидаемый предел (1/2)')
plt.title("График последовательности sqrt(n^2 + n) - n")
plt.xlabel("n")
plt.ylabel("Значение последовательности")
plt.legend()
plt.grid(True)
plt.show()
```

Значения последовательности для больших n должны стремиться к 1/2.

График показывает приближение последовательности к этому пределу.
