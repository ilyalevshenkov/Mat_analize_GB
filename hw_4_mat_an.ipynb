## 1) Найти области определения функций:   f(x) = ln(x + 2).

```python
import numpy as np
import matplotlib.pyplot as plt

# Определяем функцию f(x)
def f(x):
    return np.log(x + 2)

# Создаем массив значений x, которые входят в область определения
x_values = np.linspace(-1.99, 10, 500)

# Вычисляем значения функции f(x)
y_values = f(x_values)

# Построим график функции
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label=r'$f(x) = \ln(x + 2)$', color='blue')

# Ограничим оси для наглядности
plt.xlim(-3, 10)
plt.ylim(-5, 5)

# Добавляем оси и сетку
plt.axvline(x=-2, color='red', linestyle='--', label=r'Граница: $x = -2$')
plt.axhline(y=0, color='black', linewidth=0.5)
plt.axvline(x=0, color='black', linewidth=0.5)

# Настройка графика
plt.title("График функции $f(x) = \ln(x + 2)$ и область определения")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)

# Отобразим график
plt.show()
```

Для того чтобы найти область определения функции
f(x)=ln(x+2), нужно определить, при каких значенияхx логарифм существует.
Логарифм определён только для положительных аргументов, то есть:
x + 2 > 0.

Решим неравенство:
x > −2 .

Ответ: Область определения функции  D (f) = [ −2 , ∞], (не включает -2).

## 2) Построить график функции:

### y = x^2 + 4x + 3

```python
#import numpy as np
#import matplotlib.pyplot as plt

# Определяем функцию y = x^2 + 4x + 3
def f(x):
    return x**2 + 4*x + 3

# Создаем массив значений x
x_values = np.linspace(-10, 5, 400)

# Вычисляем значения функции f(x)
y_values = f(x_values)

# Построим график функции
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label=r'$y = x^2 + 4x + 3$', color='blue')

# Добавляем оси и сетку
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True)

# Настройки графика
plt.title("График функции $y = x^2 + 4x + 3$")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()

# Отображаем график
plt.show()
```

### y = | {x} - 1 |


```python
import numpy as np
import matplotlib.pyplot as plt

# Определяем функцию для вычисления дробной части числа
def fractional_part(x):
    return x - np.floor(x)

# Определяем функцию y = |{x} - 1|
def f(x):
    return np.abs(fractional_part(x) - 1)

# Создаем массив значений x
x_values = np.linspace(-10, 10, 1000)

# Вычисляем значения функции f(x)
y_values = f(x_values)

# Построим график функции
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label=r'$y = |\{x\} - 1|$', color='blue')

# Добавляем оси и сетку
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True)

# Настройки графика
plt.title("График функции $y = |\{x\} - 1|$")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()

# Отображаем график
plt.show()
```

## 3) Найти пределы:

### lim x → 5 [(x^2 -6x +5) / ( x^2 - 25)]

```python
import sympy as sp

# Определим переменную x
x = sp.Symbol('x')

# Определим функцию f(x) = (x^2 - 6x + 5) / (x^2 - 25)
numerator = x**2 - 6*x + 5
denominator = x**2 - 25
f_x = numerator / denominator

# Вычислим предел при x -> 5
limit_value = sp.limit(f_x, x, 5)

# Выведем результат
print(f'Предел функции при x -> 5: {limit_value}')
```

### lim x → -1 [(x^3+x+2) / (x^3 + 1)]

```python
import sympy as sp

# Определим переменную x
x = sp.Symbol('x')

# Определим функцию f(x) = (x^3 + x + 2) / (x^3 + 1)
numerator = x**3 + x + 2
denominator = x**3 + 1
f_x = numerator / denominator

# Вычислим предел при x -> -1
limit_value = sp.limit(f_x, x, -1)

# Выведем результат
print(f'Предел функции при x -> -1: {limit_value}')
```

### lim x → 3 [(2x+3)^0^5 - 3 ) / ( x - 2 )^0,5  - 1)]

```python
# import sympy as sp

# Определяем переменную x
x = sp.Symbol('x')

# Определяем функцию f(x) = (sqrt(2x + 3) - 3) / (sqrt(x - 2) - 1)
numerator = sp.sqrt(2*x + 3) - 3
denominator = sp.sqrt(x - 2) - 1
f_x = numerator / denominator

# Вычислим предел при x -> 3
limit_value = sp.limit(f_x, x, 3)

# Выведем результат
print(f'Предел функции при x -> 3: {limit_value}')
```

### lim x → 0 [( 1 - cos x) / x^2]

```python
# Определяем переменную x
x = sp.Symbol('x')

# Определяем функцию f(x) = (1 - cos(x)) / x^2
f_x = (1 - sp.cos(x)) / x**2

# Вычислим предел при x -> 0
limit_value = sp.limit(f_x, x, 0)

# Выведем результат
print(f'Предел функции при x -> 0: {limit_value}')
```

### lim x → 0  x · ctg x

### lim x → 0 (1 + 3x)^(1/2*x)

```python
import sympy as sp

# Определим переменную x
x = sp.Symbol('x')

# Определим выражение sqrt[2x]{1 - 3*x}
expr = (1 - 3*x)**(1/(2*x))

# Найдем предел выражения при x -> 0
limit_value = sp.limit(expr, x, 0)

# Выведем результат
print(f'Предел функции при x -> 0: {limit_value}')
```
