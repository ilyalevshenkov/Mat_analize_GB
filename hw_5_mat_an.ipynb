##1. limx→∞[ln(x + 3) − ln x];



```python
import sympy as sp

# Определяем переменную
x = sp.Symbol('x')

# Определяем выражение
expr = sp.log(x + 3) - sp.log(x)

# Находим предел при x -> ∞
limit_value = sp.limit(expr, x, sp.oo)

# Выводим результат
print(f'Предел при x -> ∞: {limit_value}')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Задаем большие значения для x
x_values = np.arange(1, 1000000, 10000)

# Вычисляем значения последовательности ln(x + 3) - ln(x)
sequence_values = np.log(x_values + 3) - np.log(x_values)

# Выводим последние 10 значений последовательности для больших x
print("Значения последовательности для больших x:")
for x, value in zip(x_values[-10:], sequence_values[-10:]):
    print(f"x = {x}, ln(x + 3) - ln(x) = {value}")

# Вычислим значение предела для самого большого x
largest_x = x_values[-1]
largest_value = np.log(largest_x + 3) - np.log(largest_x)
print(f"\nЗначение последовательности для самого большого x ({largest_x}): {largest_value}")

# Построим график
plt.figure(figsize=(10, 6))
plt.plot(x_values, sequence_values, color='blue', label=r'$\ln(x + 3) - \ln(x)$')
plt.axhline(y=0, color='red', linestyle='--', label='Предел = 0')
plt.title("График последовательности ln(x + 3) - ln(x)")
plt.xlabel("x")
plt.ylabel("Значение последовательности")
plt.legend()
plt.grid(True)
plt.show()
```

Используем большие значения х от 1 до 1,000,000 с шагом 10,000, чтобы наблюдать, как последовательность изменяется с увеличением x.

Расчет последовательности:

Последовательность вычисляется как
ln ( x + 3 ) − ln ( x ) ln(x+3)−ln(x).


Вывод значений: Печатаются последние 10 значений последовательности для проверки приближения к пределу 0.

##Результат:##
Значения последовательности быстро приближаются к 0 с увеличением x.

## 2. limx→0 [ln(1+2x) / arcsin 3x]


```python
import sympy as sp

# Определим переменную x
x = sp.Symbol('x')

# Определим выражение
expr = sp.log(1 + 2*x) / sp.asin(3*x)

# Найдем предел при x -> 0
limit_value = sp.limit(expr, x, 0)

# Выведем результат
print(f'Предел при x -> 0: {limit_value}')
```

```python
#import numpy as np
#import matplotlib.pyplot as plt

# Функции ln(1 + 2x) и arcsin(3x)
def f(x):
    return np.log(1 + 2 * x)

def g(x):
    return np.arcsin(3 * x)

# Задаем значения x, приближающиеся к нулю с положительной и отрицательной стороны
x_values = np.linspace(-0.0001, 0.0001, 1000)

# Вычисляем значения последовательности f(x) / g(x)
sequence_values = f(x_values) / g(x_values)

# Печатаем последние несколько значений, приближающихся к нулю
print("Значения последовательности для x, близких к 0:")
for x, value in zip(x_values[-10:], sequence_values[-10:]):
    print(f"x = {x}, f(x) / g(x) = {value}")

# Построим график для визуализации поведения функции при x -> 0
plt.figure(figsize=(10, 6))
plt.plot(x_values, sequence_values, label=r'$\frac{\ln(1+2x)}{\arcsin(3x)}$', color='blue')
plt.axhline(y=2/3, color='red', linestyle='--', label='Предел = 2/3')
plt.xlabel("x")
plt.ylabel("f(x) / g(x)")
plt.title("График последовательности ln(1+2x) / arcsin(3x)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#import numpy as np

# Определим функцию для выражения
def f(x, a):
    return ((x + a)**3 - x**3) / a

# Задаем значения x и малые значения a, приближающиеся к нулю
x = 2  # Можно выбрать любое значение для x
a_values = np.linspace(-0.01, 0.01, 1000)

# Вычисляем значения функции при разных a
sequence_values = f(x, a_values)

# Печатаем значения функции для малых значений a
print("Значения функции для малых значений a:")
for a, value in zip(a_values[-10:], sequence_values[-10:]):
    print(f"a = {a}, f(x, a) = {value}")

# Построим график для визуализации поведения функции при a -> 0
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(a_values, sequence_values, label=r'$\frac{(x + a)^3 - x^3}{a}$', color='blue')
plt.axhline(y=3*x**2, color='red', linestyle='--', label=r'Предел = $3x^2$')
plt.xlabel("a")
plt.ylabel("f(x, a)")
plt.title(f"График функции для x = {x}")
plt.legend()
plt.grid(True)
plt.show()
```

Раскроем куб в числителе и подставим в исходное выражение:

[х^3 + 3x^2 * a + 3x * a^2 + a^3 - x^3]/ a

Убираем х^3 и делим каждый член на (а):

3x^2 +3xa + a^2

При a → 0, остаются только члены, не зависящие от a

## Ответ: 3х^2

## 2. Установить характер разрыва функции в точке x0:

### 2.1. f(x) = (x^2 - 16x) / (x + 4) , при x0 = - 4;

```python
import sympy as sp

# Определение переменной
x = sp.Symbol('x')

# Определение функции
numerator = x**2 - 16*x
denominator = x + 4
f = numerator / denominator

# Определение точки разрыва
x0 = -4

# Упрощение функции для анализа
simplified_f = sp.simplify(f)

# Вычисление пределов при x стремящемся к x0 с обеих сторон
limit_left = sp.limit(f, x, x0, dir='-')
limit_right = sp.limit(f, x, x0, dir='+')

# Вывод результатов
print("Упрощенная функция:")
print(simplified_f)

print(f"\nПредел функции при x стремящемся к {x0} слева:")
print(limit_left)

print(f"\nПредел функции при x стремящемся к {x0} справа:")
print(limit_right)

# Анализ характера разрыва
if limit_left == sp.oo or limit_left == -sp.oo or limit_right == sp.oo or limit_right == -sp.oo:
    print(f"\nФункция имеет бесконечный разрыв в точке x0 = {x0}.")
elif limit_left == limit_right:
    print(f"\nФункция имеет удаляемый разрыв в точке x0 = {x0}.")
else:
    print(f"\nФункция имеет скачкообразный разрыв в точке x0 = {x0}.")
```

### 2.2 f(x) = sin x / x,  при x0 = 0.

```python
import sympy as sp

# Определение переменной
x = sp.Symbol('x')

# Определение функции
f = sp.sin(x) / x

# Определение точки разрыва
x0 = 0

# Вычисление предела функции при x стремящемся к x0
limit = sp.limit(f, x, x0)

# Вывод результата
print(f"Предел функции f(x) = sin(x) / x при x стремящемся к {x0}:")
print(limit)

# Анализ характера разрыва
if limit == sp.oo or limit == -sp.oo:
    print(f"Функция имеет бесконечный разрыв в точке x = {x0}.")
elif limit.is_real:
    print(f"Функция имеет удаляемый разрыв в точке x = {x0}, и предел функции равен {limit}.")
else:
    print(f"Функция имеет скачкообразный разрыв в точке x = {x0}.")
```

## 3. Исследовать на непрерывность функцию f(x) в точке x0:

### 3.1. f(x) = arctg[2 / (x - 1)],  при x0 = 1

```python
#import sympy as sp

# Определение переменной
x = sp.Symbol('x')

# Определение функции
f = sp.atan(2 / (x - 1))

# Определение точки разрыва
x0 = 1

# Вычисление предела функции при x стремящемся к x0 с обеих сторон
limit_left = sp.limit(f, x, x0, dir='-')
limit_right = sp.limit(f, x, x0, dir='+')

# Вывод результатов
print(f"Предел функции f(x) = arctan(2 / (x - 1)) при x стремящемся к {x0} слева:")
print(limit_left)

print(f"Предел функции f(x) = arctan(2 / (x - 1)) при x стремящемся к {x0} справа:")
print(limit_right)

# Анализ характера разрыва
if limit_left == limit_right:
    print(f"Функция имеет разрыв в точке x = {x0}.")
    print(f"Тип разрыва: {'бесконечный' if limit_left in [sp.oo, -sp.oo] else 'удаляемый или скачкообразный'}")
else:
    print(f"Функция имеет скачкообразный разрыв в точке x = {x0}.")
```
