## Задание №1 Найти интервалы возрастания и убывания функций:

## 1) f(x) = x + e^(−x)

```python
import numpy as np
import matplotlib.pyplot as plt

# Определяем функцию и её производную
def f(x):
    return x + np.exp(-x)

def df(x):
    return 1 - np.exp(-x)

# Определение критической точки (где производная равна нулю)
critical_point = 0  # Решение уравнения 1 - e^(-x) = 0

# Определение интервалов возрастания и убывания
intervals = {
    "убывания": (-np.inf, critical_point),
    "возрастания": (critical_point, np.inf)
}

# Печать интервалов возрастания и убывания
print("Интервалы возрастания и убывания функции:")
for key, value in intervals.items():
    print(f"Интервал {key}: {value}")

# Создаем массив x от -2 до 2 для построения графика
x = np.linspace(-2, 2, 400)

# Вычисляем значения функции и её производной
y = f(x)
dy = df(x)

# Настройка графика
plt.figure(figsize=(10, 6))

# График функции
plt.plot(x, y, label=r'$f(x) = x + e^{-x}$', color='blue')

# Линия x = 0
plt.axvline(0, color='grey', linestyle='--')

# Отметка критической точки
plt.scatter(0, f(0), color='red', zorder=5, label='Критическая точка (0, 1)')

# Закраска интервалов убывания и возрастания
plt.fill_between(x, y, where=(x < critical_point), color='orange', alpha=0.3, label='Убывание')
plt.fill_between(x, y, where=(x > critical_point), color='green', alpha=0.3, label='Возрастание')


plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('График функции f(x) = x + e^{-x} и её интервалы возрастания/убывания')
plt.legend()
plt.grid(True)
plt.show()
```

## Задача 4. Найти асимптоты графиков функций:

## y = 3x / [x+2]

```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

# Определяем переменную и функцию
x = sp.symbols('x')
y = 3 * x / (x + 2)

# Найдем вертикальные асимптоты
# Вертикальная асимптота возникает, когда знаменатель равен нулю
vertical_asymptote = sp.solve(x + 2, x)

# Найдем горизонтальные асимптоты
# Горизонтальная асимптота определяется пределом функции при x стремящемся к бесконечности
horizontal_asymptote = sp.limit(y, x, sp.oo)

# Определяем функцию для построения графика
def func(x):
    return 3 * x / (x + 2)

# Определяем диапазон значений для x
x_vals = np.linspace(-10, 10, 400)
y_vals = func(x_vals)

# Настраиваем график
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='y = 3x / (x + 2)', color='blue')

# Добавляем вертикальные асимптоты
for va in vertical_asymptote:
    plt.axvline(x=va, color='red', linestyle='--', label=f'Вертикальная асимптота: x = {va}')

# Добавляем горизонтальную асимптоту
plt.axhline(y=float(horizontal_asymptote), color='green', linestyle='--', label=f'Горизонтальная асимптота: y = {horizontal_asymptote}')

# Настраиваем график
plt.title('График функции с асимптотами')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.ylim(-10, 10)  # Ограничиваем диапазон по y для лучшего отображения

# Показываем график
plt.show()

print(f"Вертикальные асимптоты: {vertical_asymptote}")
print(f"Горизонтальная асимптота: y = {horizontal_asymptote}")
```

## Задача 5. Провести полное исследование и построить графики функций:

## 1) y = ln(1 − x^2)

```python
# Определяем переменную и функцию
x = sp.Symbol('x')
y = sp.ln(1 - x**2)

# Область определения функции
domain = (-1, 1)  # Из неравенства 1 - x^2 > 0

# Точки пересечения с осями
y_intercept = y.subs(x, 0)
x_intercepts = sp.solve(sp.Eq(y, 0), x)

# Вертикальные асимптоты
vertical_asymptotes = sp.solve(sp.Eq(1 - x**2, 0), x)

# Первая производная и критические точки
y_prime = sp.diff(y, x)
critical_points = [p for p in sp.solve(y_prime, x) if domain[0] < p < domain[1]]

# Вывод результатов
print(f"Область определения функции: {domain}")
print(f"Точка пересечения с осью Oy: (0, {y_intercept})")
print(f"Точка пересечения с осью Ox: {x_intercepts}")
print(f"Вертикальные асимптоты: x = {vertical_asymptotes}")
print(f"Первая производная: {y_prime}")
print(f"Критические точки: {critical_points}")

# Определяем интервалы возрастания и убывания
for i in range(len(critical_points) + 1):
    interval = (domain[0], critical_points[0]) if i == 0 else (critical_points[-1], domain[1]) if i == len(critical_points) else (critical_points[i-1], critical_points[i])
    test_point = (interval[0] + interval[1]) / 2
    derivative_value = y_prime.subs(x, test_point)
    print(f"Функция {'возрастает' if derivative_value > 0 else 'убывает'} на интервале: {interval}")

# Построение графика функции
x_vals = np.linspace(-0.99, 0.99, 400)
y_vals = np.log(1 - x_vals**2)
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='y = ln(1 - x^2)', color='blue')
plt.axvline(x=-1, color='red', linestyle='--', label='Вертикальная асимптота: x = -1')
plt.axvline(x=1, color='red', linestyle='--', label='Вертикальная асимптота: x = 1')
plt.title('График функции y = ln(1 - x^2) с асимптотами')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Определяем переменную и функцию
x = sp.Symbol('x')
y = x**2 * sp.exp(-x)

# Область определения функции: вся числовая прямая (R)
domain = (-sp.oo, sp.oo)

# Точки пересечения с осями
y_intercept = y.subs(x, 0)
x_intercepts = sp.solve(sp.Eq(y, 0), x)

# Производная и критические точки
y_prime = sp.diff(y, x)
critical_points = sp.solve(sp.Eq(y_prime, 0), x)

# Вторая производная и точки перегиба
y_double_prime = sp.diff(y_prime, x)
inflection_points = sp.solve(sp.Eq(y_double_prime, 0), x)

# Горизонтальная асимптота
horizontal_asymptote = 0  # y -> 0 при x -> ∞

# Вывод результатов
print(f"Область определения функции: R")
print(f"Точка пересечения с осью Oy: (0, {y_intercept})")
print(f"Точки пересечения с осью Ox: {x_intercepts}")
print(f"Критические точки: {critical_points}")
print(f"Точки перегиба: {inflection_points}")
print(f"Горизонтальная асимптота: y = {horizontal_asymptote}")

# Построение графика функции
x_vals = np.linspace(-1, 8, 400)
y_vals = x_vals**2 * np.exp(-x_vals)
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label=r'$y = x^2 e^{-x}$', color='blue')
plt.axhline(y=0, color='green', linestyle='--', label='Горизонтальная асимптота: y = 0')
plt.title('График функции y = x^2 * e^(-x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

## 4) y = x − ln x

```python
# Определяем переменную и функцию
x = sp.Symbol('x', real=True, positive=True)  # x > 0 для ln(x)
y = x - sp.ln(x)

# Область определения функции: x > 0
domain = (0, sp.oo)

# Точки пересечения с осями
y_intercept = None  # Нет пересечения с Oy, так как x > 0
x_intercepts = sp.solve(sp.Eq(y, 0), x)

# Производная и критические точки
y_prime = sp.diff(y, x)
critical_points = sp.solve(sp.Eq(y_prime, 0), x)
critical_points_coords = [(p.evalf(), (p - sp.ln(p)).evalf()) for p in critical_points]

# Вторая производная и точки перегиба
y_double_prime = sp.diff(y_prime, x)
inflection_points = sp.solve(sp.Eq(y_double_prime, 0), x)
inflection_points_coords = [(p.evalf(), (p - sp.ln(p)).evalf()) for p in inflection_points]

# Горизонтальная асимптота
horizontal_asymptote = None  # Нет горизонтальных асимптот, так как y -> ∞ при x -> ∞

# Вывод результатов
print(f"Область определения функции: x > 0")
print(f"Точка пересечения с осью Oy: {y_intercept}")
print(f"Точки пересечения с осью Ox: {x_intercepts}")
print(f"Критические точки: {critical_points_coords}")
print(f"Точки перегиба: {inflection_points_coords}")
print(f"Горизонтальная асимптота: {horizontal_asymptote}")

# Построение графика функции
x_vals = np.linspace(0.1, 5, 400)
y_vals = x_vals - np.log(x_vals)
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label=r'$y = x - \ln(x)$', color='blue')

# Отображение критических точек и точек перегиба
for cp in critical_points_coords:
    plt.plot(cp[0], cp[1], 'ro')  # Критические точки
    plt.text(cp[0], cp[1], f'({cp[0]:.2f}, {cp[1]:.2f})', fontsize=9, ha='right')

for ip in inflection_points_coords:
    plt.plot(ip[0], ip[1], 'go')  # Точки перегиба
    plt.text(ip[0], ip[1], f'({ip[0]:.2f}, {ip[1]:.2f})', fontsize=9, ha='left')

plt.title('График функции y = x - ln(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```
