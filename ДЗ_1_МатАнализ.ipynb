1. Представьте в виде несократимой рациональной дроби:
а) 0.(216);
б) 1.0(01).

```python
from fractions import Fraction

def convert_periodic_to_fraction(non_periodic_part, periodic_part):
    # Длина непериодической части и периодической части
    len_non_periodic = len(non_periodic_part)
    len_periodic = len(periodic_part)

    # Создаем числитель и знаменатель
    if len_non_periodic > 0:
        non_periodic_value = int(non_periodic_part)
        numerator = int(non_periodic_part + periodic_part) - non_periodic_value
        denominator = 10 ** len_periodic - 1
        denominator *= 10 ** len_non_periodic
    else:
        numerator = int(periodic_part)
        denominator = 10 ** len_periodic - 1

    # Возвращаем дробь в несократимой форме
    return Fraction(numerator, denominator)

# Пример 1: 0.(216)
non_periodic_part_1 = ''  # Непериодическая часть (перед запятой ничего нет)
periodic_part_1 = '216'  # Периодическая часть
fraction_1 = convert_periodic_to_fraction(non_periodic_part_1, periodic_part_1)

# Пример 2: 1.0(01)
non_periodic_part_2 = '1'  # Непериодическая часть перед периодом
periodic_part_2 = '01'  # Периодическая часть
fraction_2 = convert_periodic_to_fraction(non_periodic_part_2, periodic_part_2)

print(f"0.(216) = {fraction_1}")
print(f"1.0(01) = {fraction_2}")
```

### 2. Проверьте любым способом, являются ли данные логические формулы тавтологией:

а) (AA ∨ BB) → (BB ∨ AA̅)

б) AA → AA ∨ (BB ∧ AA)

```python
import itertools

def is_tautology(formula):

    # Генерация всех возможных комбинаций значений логических переменных (A и B)
    variables = ['A', 'B']
    values = list(itertools.product([True, False], repeat=len(variables)))

    # Проверка каждой комбинации
    for value_set in values:
        # Сопоставляем переменные и их значения
        A, B = value_set

        # Используем eval для вычисления формулы
        if not eval(formula):
            return False  # Если хотя бы одна комбинация не удовлетворяет, формула не является тавтологией

    return True  # Если все комбинации удовлетворяют, это тавтология

# Формулы для проверки
# Импликация (A -> B) эквивалентна (not A or B)
formula_1 = '(not A or B) or (B or not A)'  # Эквивалентная формула импликации для (A or B) -> (B or not A)
formula_2 = 'not A or (A or (B and A))'  # Эквивалентная формула для A -> (A or (B and A))

# Проверка формул
result_1 = is_tautology(formula_1)
result_2 = is_tautology(formula_2)

print(f"Формула (A ∨ B) → (B ∨ ¬A) является тавтологией: {result_1}")
print(f"Формула A → A ∨ (B ∧ A) является тавтологией: {result_2}")
```

### 3. Сформулируйте словесно высказывания:

а) (AA̅ ∨ BB) → CC̅

б) CC → (AA ∨ BB)

где А: сегодня светит солнце; В: сегодня сыро; С: я поеду на дачу.

```python
import itertools

# Функция для проверки, является ли формула тавтологией
def is_tautology(formula):
    # Генерация всех возможных комбинаций значений логических переменных (A, B, C)
    variables = ['A', 'B', 'C']
    values = list(itertools.product([True, False], repeat=len(variables)))

    # Проверка каждой комбинации
    for value_set in values:
        # Сопоставляем переменные и их значения
        A, B, C = value_set

        # Используем eval для вычисления формулы
        if not eval(formula):
            return False  # Если хотя бы одна комбинация не удовлетворяет, формула не является тавтологией

    return True  # Если все комбинации удовлетворяют, это тавтология

# Словесное описание формул
def describe_formula(formula, A_desc, B_desc, C_desc):
    if formula == 'a':
        return f"Если {A_desc} или {B_desc}, то {C_desc} не выполнится."
    elif formula == 'b':
        return f"Если {C_desc}, то либо {A_desc}, либо {B_desc} выполнится."
    else:
        return "Неизвестная формула."

# Формулы для проверки
# Формула 1: (¬A ∨ B) → ¬C эквивалентна (not A or B) → not C
# Импликация (P → Q) эквивалентна (not P or Q)
formula_1 = '(not (not A or B) or not C)'  # Эквивалентная формула импликации
formula_2 = '(not C or (A or B))'  # Эквивалентная формула импликации

# Проверка тавтологичности формул
result_1 = is_tautology(formula_1)
result_2 = is_tautology(formula_2)

# Описание переменных
A_description = "сегодня светит солнце"
B_description = "сегодня сыро"
C_description = "я поеду на дачу"

# Вывод результатов
print(f"Формула (¬A ∨ B) → ¬C является тавтологией: {result_1}")
print(describe_formula('a', A_description, B_description, C_description))

print(f"Формула C → (A ∨ B) является тавтологией: {result_2}")
print(describe_formula('b', A_description, B_description, C_description))
```
